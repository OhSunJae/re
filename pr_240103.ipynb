```python
class Hello:
    def __enter__(self):
        print('enter....')
        return self
    
    def sayHello(self, name):
        print('hello ' + name)
        
    def __exit__(self, exc_type, exc_val, exc_tb):
        print('exit...')
```

```python
with Hello() as h:
    h.sayHello('obama')
    h.sayHello('trump')
```

```python
class Oper:
    def __init__(self):
        self.res = 0 
        
    def plus(self, num):
        self.res += num
        return self.res
    
student1 = Oper()
student2 = Oper()
```

```python
print(student1.plus(3))
print(student2.plus(4))
print(student1.plus(5))
```
